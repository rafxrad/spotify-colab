```python
import pandas as pd
from google.colab import drive
from google.colab import files
import requests
import numpy
import os
```

```python

drive.mount("/content/gdrive")
```

```python
local_dir = '/content/gdrive/MyDrive/data'
```

Carregando todos os arquivos json.

```python
json_files = [f for f in os.listdir(local_dir) if f.endswith(".json")]
print(json_files)
```

Transformando todos os arquivos em dataframe.


```python
dfs = []
for file in json_files:
  df = pd.read_json('{}/{}'.format(local_dir,file))
  dfs.append(df)

```

Checando a estrutura das colunas antes de combiná-las.

```python
for num, df in enumerate(dfs):
  print(f"index: {num}': 'df':{df.columns}")
```

```python
concat_dfs = pd.concat(dfs)
```

```python
concat_dfs
```

```python
columns_dispose = ['ip_addr_decrypted', 'user_agent_decrypted', 'episode_name', 'episode_show_name', 'spotify_episode_uri', 'spotify_track_uri']
for column in columns_dispose:
  concat_dfs = concat_dfs.drop(column, axis=1)

print('Colunas deletadas com sucesso.')
```

```python
concat_dfs
```

Renomeando as colunas.

```python
print(concat_dfs.columns)
```

```python
concat_dfs.rename(
columns= {
           "ts": "timestamp",
           "username": "user",
           "ms_played": "minutes played",
           "conn_country": "country connected",
           "user agent": "user_agent_decrypted",
           "master_metadata_track_name": "track name",
           "master_metadata_album_artist_name": "artist name",
           "master_metadata_album_album_name": "album name"
           },
          inplace=True)
```
